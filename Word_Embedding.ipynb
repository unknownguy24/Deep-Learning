```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
### Vocabulary size
voc_size=10000   #tells us the size of the dictionary which will tell us the index of the word
```

```python
#One Hot Representation
onehot_repr=[one_hot(words,voc_size)for words in sent] #gives an index to each word
print(onehot_repr)
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences  #makes sure the size i.e number of words in the sentence is same
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_length=8   #maximum sentence size setting
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length) #as max size is 8, padding is applied as pre meaning out of 8 words if 4 are present before the 4 all will have 0 index
print(embedded_docs)
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length)) #we have taken 10 dimensions(features)
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_docs)) #to see how our words are converted to vectors i.e the Feature Representation
```

```python
embedded_docs[0] #before passing through embeded layer
```

```python
print(model.predict(embedded_docs)[0])   #each word having 10 dimension and creating a vector
```
